```python
import pygame
import tkinter as tk
from tkinter import filedialog
import os

# Initialize pygame mixer for playing music
pygame.mixer.init()

# Global variables to hold the playlist and the current song index
playlist = []
current_song_index = 0

# Define a function to play music from the playlist
def play_music():
    if playlist:
        pygame.mixer.music.load(playlist[current_song_index])
        pygame.mixer.music.play()
        status_label.config(text=f"Playing: {playlist[current_song_index]}")
    else:
        status_label.config(text="No songs in the playlist.")

# Define a function to stop music
def stop_music():
    pygame.mixer.music.stop()
    status_label.config(text="Stopped")

# Define a function to pause music
def pause_music():
    pygame.mixer.music.pause()
    status_label.config(text="Paused")

# Define a function to unpause music
def unpause_music():
    pygame.mixer.music.unpause()
    status_label.config(text="Playing...")

# Define a function to load all MP3 files from a selected folder
def load_music_from_folder():
    global playlist
    folder = filedialog.askdirectory()  # Let the user choose a folder
    if folder:
        playlist = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith('.mp3')]
        if playlist:
            update_playlist_display()
            play_music()
            status_label.config(text="Loaded music from folder.")
        else:
            status_label.config(text="No MP3 files found in the folder.")

# Define a function to move to the next song in the playlist
def next_song():
    global current_song_index
    if len(playlist) > 0:
        current_song_index = (current_song_index + 1) % len(playlist)
        play_music()

# Define a function to move to the previous song in the playlist
def previous_song():
    global current_song_index
    if len(playlist) > 0:
        current_song_index = (current_song_index - 1) % len(playlist)
        play_music()

# Define a function to update the playlist display
def update_playlist_display():
    playlist_box.delete(0, tk.END)
    for song in playlist:
        playlist_box.insert(tk.END, os.path.basename(song))

# Check if the music is still playing, if not move to the next song
def check_music_status():
    if not pygame.mixer.music.get_busy():  # Check if the current song has finished
        next_song()
    root.after(1000, check_music_status)  # Check the music status every second

# Create the main window using tkinter
root = tk.Tk()
root.title("Music Player with Playlist")
root.geometry("500x400")

# Label for status
status_label = tk.Label(root, text="No music loaded", width=40)
status_label.pack(pady=10)

# Playlist display
playlist_box = tk.Listbox(root, height=10, width=50)
playlist_box.pack(pady=10)

# Play button
play_button = tk.Button(root, text="Play", width=15, command=play_music)
play_button.pack(pady=5)

# Pause button
pause_button = tk.Button(root, text="Pause", width=15, command=pause_music)
pause_button.pack(pady=5)

# Unpause button
unpause_button = tk.Button(root, text="Unpause", width=15, command=unpause_music)
unpause_button.pack(pady=5)

# Stop button
stop_button = tk.Button(root, text="Stop", width=15, command=stop_music)
stop_button.pack(pady=5)

# Load button
load_button = tk.Button(root, text="Load Music Folder", width=20, command=load_music_from_folder)
load_button.pack(pady=5)

# Next song button
next_button = tk.Button(root, text="Next", width=15, command=next_song)
next_button.pack(pady=5)

# Previous song button
previous_button = tk.Button(root, text="Previous", width=15, command=previous_song)
previous_button.pack(pady=5)

# Start checking the music status
check_music_status()

# Run the tkinter main loop
root.mainloop()
```
